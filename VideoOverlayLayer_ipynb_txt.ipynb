```python
#Pulling Video from googled drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#set video as object to be displayed in map
from base64 import b64encode
mp4 = open('/content/drive/MyDrive/data/WYE02.mp4','rb').read()
decoded_vid = "data:video/mp4;base64," + b64encode(mp4).decode()
```

```python
#display the video to notebook
from IPython.display import HTML
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML("""
<video width=200 controls autoplay>
      <source src="%s" type="video/mp4">
</video>
""" % data_url)
```

```python
import folium
m = folium.Map(location=[47.6062, -122.3321], zoom_start=22, zoom_control=True)
video = folium.raster_layers.VideoOverlay( video_url=decoded_vid,  bounds=[[47.6061, -122.322], [47.6063, -122.345]], opacity=0.85,  attr="Sewer Inspection",  autoplay=True, loop=False,)
video.add_to(m)

# add in functionality to annotate the map
from folium.plugins import Draw
draw = Draw(export=True)
draw.add_to(m)
m
```
